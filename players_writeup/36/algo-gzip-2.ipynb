```python
class Bitfield:
    def __init__(self, 
                 bit_string: bytes,
                 remaining_bits: int = 0,
        ):
        self.bytes_string = bit_string
        self.bytes_idx = 0
        
        self.bits = remaining_bits
        self.bitfield = 0x0
        self.position = 0  # Tracks current read position in the bit string

    def needbits(self, n):
        # Ensure there are at least n bits left in the bitfield
        while self.bits < n:
            self._more()

    def _mask(self, n):
        # Create a mask for n bits
        return (1 << n) - 1

    def tell(self):
        # Return the current byte and bit positions
        return self.position // 8, self.position % 8

    def tellbits(self):
        # Return the total number of bits read
        return self.position

    def _more(self):
        _rd = self.bytes_string[self.bytes_idx]  # this is an int
        self.bytes_idx += 1
        print ("old_bitfield", format(self.bitfield, '08b'), "old_bits", self.bits, "read", format(c, '08b'))
        _t = _rd << self.bits
        self.bitfield += _t
        self.bits += 8
        print ("more", "at", self.tell(), "read", format(c), '08b', "set_bitfield", format(self.bitfield, '08b'))

    def readbits(self, n = 8):
        if n > self.bits:
            self.needbits(n)
        # Read exactly n bits from the bit string and return them as an integer
        self.needbits(n)
        r = self.bitfield & self._mask(n)
        self.bits -= n
        self.bitfield >>= n
        print("readbits", "return", format(r, '08b'), "bitfield", format(self.bitfield, '08b'))
        return r
```

```python
_test = b'[What can I say? Mamba out! --KobeBryant]XXXX'

def _decode_payload(payload: bytes, bits: int, remaining_bits: int = 0, code_to_char: dict = None):
    _dec = Bitfield(payload, remaining_bits)
    _text = ''
    try:
        for _ in range(1000):
            r = _dec.readbits(bits)
            zo_r = format(r, '08b')[2:]
            try:
                code = code_to_char[zo_r]
            except KeyError:
                code = b'\xFF'
            print(f"read {zo_r} -> {code}")
            if code == b'\xFF':
                break
            _text += code
    except IndexError:
        pass
    return _text
        
_decode_payload(_test, 6, 1, _code_to_char)
```

```python
import random

def shuffle_under_seed(ls, seed):
  # Shuffle the list ls using the seed `seed`
  random.seed(seed)
  random.shuffle(ls)
  return ls

def unshuffle_list(shuffled_ls, seed):
  n = len(shuffled_ls)
  # Perm is [1, 2, ..., n]
  perm = [i for i in range(1, n + 1)]
  # Apply sigma to perm
  shuffled_perm = shuffle_under_seed(perm, seed)
  # Zip and unshuffle
  zipped_ls = list(zip(shuffled_ls, shuffled_perm))
  zipped_ls.sort(key=lambda x: x[1])
  return [a for (a, b) in zipped_ls]

print(unshuffle_list(shuffle_under_seed([1, 2, 3, 4, 5], 123423), 123423))

def undo_mixup(shuffled_text, seed):
    if isinstance(shuffled_text, str):
        shuffled_text = [ord(c) for c in shuffled_text]
    original_text = unshuffle_list(shuffled_text, seed)
    original_text = ''.join(chr(c ^ 17) for c in original_text)
    
    return original_text

SEED = 10086

def do_mixup(text: str, SEED):
    text = [ord(c) ^ 17 for c in text]
    random.seed(SEED)
    random.shuffle(text)
    return text
    
print(do_mixup('ABCDEFGHI', SEED))
undo_mixup(do_mixup('ABCDEFGHI', SEED), SEED)
```
