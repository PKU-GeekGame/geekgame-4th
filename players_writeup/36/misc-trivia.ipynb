> 在全新安装的 Ubuntu Desktop 22.04 系统中，把音量从 75% 调整到 25% 会使声音减小多少分贝？（保留一位小数）

这个问题是最坑的，第一次提交就寄了。每一个操作系统和发行版对于多少音量对应多少分贝都有自己的算法。Ubuntu是开源的Linux发行版，理论上可以通过看源码解决；但是我懒了，直接装了个虚拟机，使用PulseAudio的`pactl`能得到分贝单位的当前音量：

```bash
test@test-virtual-machine:~$ pactl set-sink-volume 0 75%
test@test-virtual-machine:~$ pactl get-sink-volume 0 
Volume: front-left: 49152 /  75% / -7.50 dB,   front-right: 49152 /  75% / -7.50 dB
        balance 0.00
test@test-virtual-machine:~$ pactl set-sink-volume 0 25%
test@test-virtual-machine:~$ pactl get-sink-volume 0 
Volume: front-left: 16384 /  25% / -36.12 dB,   front-right: 16384 /  25% / -36.12 dB
        balance 0.00
```

**答案是：28.6**


> 这张照片用红框圈出了一个建筑。离它最近的已开通地铁站是什么？

对于OSINT题，首先我们需要知道这张照片的拍摄位置大概在哪。
把图片中的所有文字输入搜索引擎，我们能查到一个楼盘名称“月亮河七星公馆”，位于北京通州。我们点开高德地图找到这个楼盘，在附近搜索“塔”，能得到附近的塔是“燃灯佛舍利塔”。然后点开高德地图，搜索附近的地铁站，答案就出来了。

**答案是：通州北关**
